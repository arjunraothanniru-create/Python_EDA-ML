```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.tree import plot_tree
from sklearn.ensemble import RandomForestClassifier
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import accuracy_score
from sklearn.model_selection import cross_val_score
```

```python
df = pd.read_excel('/content/glass_values.xlsx')
```

```python
df.to_csv('/content/glass_values.xlsx', index = False)
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
x = df.drop('Type',axis = 1)
y = df['Type']
```

```python
x
```

```python
y
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
rf = RandomForestClassifier(random_state=42,n_estimators=100)
```

```python
rf.fit(x_train, y_train)
```

```python
y_pred =rf.predict(x_test)
```

```python
y_test
```

```python
y_pred
```

```python
print("Accuracy:", accuracy_score(y_test, y_pred))
```

```python
#hyper parameter tunning
cross_val_score(RandomForestClassifier(), x, y,scoring ='accuracy', cv=5).mean()
```

```python
#n_estimators

cross_val_score(RandomForestClassifier(n_estimators=4), x, y,scoring ='accuracy', cv=5).mean()
```

```python
#max_depth
cross_val_score(RandomForestClassifier(max_depth=4), x, y,scoring ='accuracy', cv=5).mean()
```

```python
#max_features
cross_val_score(RandomForestClassifier(max_features=4), x, y,scoring ='accuracy', cv=5).mean()
```

```python
#min_samples_split
cross_val_score(RandomForestClassifier(min_samples_split=4), x, y,scoring ='accuracy', cv=5).mean()
```

```python
#using gridsearchcv
params1 = {'min_samples_leaf': [None , 2,3,4,5,6,7,8]}
grid1 = GridSearchCV(RandomForestClassifier(),params1)
grid1.fit(x,y)
```

```python
grid1.best_params_
```

```python
grid1.best_score_
```

```python
params2 = {'max_depth': [None , 2,3,4,5,6,7,8]}
grid2 = GridSearchCV(RandomForestClassifier(),params2)
grid2.fit(x,y)
```

```python
grid2.best_score_
```

```python
params3 = {'max_features': [ 2,3,4,5,6,7,8]}
grid3 = GridSearchCV(RandomForestClassifier(),params3)
grid3.fit(x,y)
```
