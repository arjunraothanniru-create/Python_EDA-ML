# DATA VISUALIZATION

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Histogram for numerical columns
numerical_columns = df.select_dtypes(include=['int', 'float']).columns
for column in numerical_columns:
    sns.histplot(df[column])
    plt.title(f'Histogram of {column}')
    plt.xlabel(column)
    plt.ylabel('Frequency')
    plt.show()
```

```python
#Boxplots for numerical columns
plt.figure(figsize=(10, 8))
sns.boxplot(data=df[numerical_columns])
plt.title('Boxplot of Numerical Columns')
plt.xlabel('Numerical Columns')
plt.ylabel('Values')
plt.xticks(rotation=45)
plt.show()
```

```python
#Bar chart analysis for categorical columns
categorical_columns = df.select_dtypes(include=['object']).columns
for column in categorical_columns:
    plt.figure(figsize=(8, 6))
    sns.countplot(data=df, x=column)
    plt.title(f'Bar Chart of {column}')
    plt.xlabel(column)
    plt.ylabel('Count')
    plt.xticks(rotation=45)
    plt.show()
```

# Standardization of Numerical Variables

```python

"""Standardization, also known as z-score normalization, is a method used to scale numerical variable
s to have a mean of 0 and a standard deviation of 1. It transforms the data distribution in such a
way that the mean of the standardized data is 0 and the standard deviation is 1. This process is
particularly useful when the features have different scales or units, making it difficult for
certain algorithms to interpret them correctly. The formula for standardization is
z-score normalization is: z= (x-mu)/sigma"""
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
#Before Standardization - Plot histograms
plt.title('Before Standardization')
for column in numerical_columns:
    sns.histplot(df[column], kde=True, label=column)
plt.legend()
plt.show()
```

```python
#Standardization
scaler = StandardScaler()
df[numerical_columns] = scaler.fit_transform(df[numerical_columns])
```

```python
#After Standardization - Plot histograms
plt.title('After Standardization')
for column in numerical_columns:
    sns.histplot(df[column], kde=True, label=column)
plt.legend()
plt.show()
```

# Conversion of Categorical Data into Dummy Variable

```python

"""Converting categorical data into dummy variables, also known as one-hot encoding, is
necessary because many machine learning algorithms cannot directly handle categorical data.
One-hot encoding transforms categorical variables into a binary format where each category
becomes a binary column (0 or 1). This encoding allows algorithms to properly interpret and
utilize categorical data without assigning any ordinal relationship between categories.
Here's how one-hot encoding works: For each unique category in a categorical column, a new
binary column is created. If an observation belongs to a particular category, the corresponding
binary column is marked as 1, and all other binary columns for that categorical variable are
marked as 0."""
```

```python
#Identifying categorical columns
categorical_columns = df.select_dtypes(include=['object']).columns
```

```python
categorical_columns
```

```python
#Applying one-hot encoding
df_encoded = pd.get_dummies(df, columns=categorical_columns, drop_first=True)
```

```python
df_encoded
```

```python
#Display a portion of the transformed dataset
print("Transformed Dataset:")
print(df_encoded.head())
```
